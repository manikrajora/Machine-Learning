```python
import numpy as np

import gym
from gym import wrappers
import matplotlib.pyplot as plt
import time
```

```python
# Simulate Environment

def sim_env(env, Q, n = 1000, max_steps = 100):
    score = []
    for i in range(n):
        s = env.reset()
        d = False
        step_n = 0
        while not d:
            step_n += 1
            a = np.argmax(Q[s,:])
            s1,r,d,_ = env.step(a)
            s = s1
            if step_n >= max_steps:
                break
        score.append(r)
    return(score)
```
