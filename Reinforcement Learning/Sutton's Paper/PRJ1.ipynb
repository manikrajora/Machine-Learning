```python
from __future__ import division
import math
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def Data_gen(n=10):
    TrainSet=[]
    States=["A","B","C","D","E","F","G"]
    startP="D"
    for i in range(n):
        curP=startP
        seq=[curP]
        cur_ind=States.index(curP)
        while cur_ind!=0 and cur_ind!=6:
            cur_ind=States.index(curP)
            rand=random.uniform(0, 1)
            if rand<=0.5:
                cur_ind+=-1
            else:
                cur_ind+=1
            curP=States[cur_ind]
            seq.append(curP)
            
        TrainSet.append(seq)
    return TrainSet
```

```python
def learn_oneset(Trainset,alpha,lamd,w):
    States=["A","B","C","D","E","F","G"]
    X=[np.array([1,0,0,0,0]),np.array([0,1,0,0,0]),np.array([0,0,1,0,0]),np.array([0,0,0,1,0]),np.array([0,0,0,0,1])]
    test=1
    P_act = np.array([1/6,2/6,3/6,4/6,5/6])

    P = np.zeros(7)
    for s in range(7):
        if s == 0:
            P[s] = 0.0
        elif s == 6:
            P[s] = 1.0
        else:
            px = X[s-1]
            P[s] = np.dot(np.transpose(px),w)

    e = 1100
    
    while e > 0.0001:
        delta_w = np.zeros(5)
        for m in range(len(Trainset)):
            Trainset_use = Trainset[m]
            for i in range(len(Trainset_use)-1):
                ind_1 = States.index(Trainset_use[i])
                ind_2 = States.index(Trainset_use[i+1])
                sum = np.zeros(5)
                for j in range(i+1):
                    sum += (lamd**(i-j))*X[States.index(Trainset_use[j])-1]
                delta_w += alpha*(P[ind_2]-P[ind_1])*sum
        
        w += delta_w
        for s in range(7):
            if s == 0:
                P[s] = 0
            elif s == 6:
                P[s] = 1
            else:
                px = X[s-1]
                P[s] = np.dot(np.transpose(px),w)
        e = np.linalg.norm(delta_w)
                
    RMSE = np.linalg.norm(P[1:6]-P_act)/np.sqrt(5)  
    return [w,P,RMSE]

RMSE_total = 0.0

lamda = np.array([0.0, 0.1, 0.3, 0.5, 0.7, 0.9, 1.0])
#lamda = np.array([0.1, 0.3, 0.5, 0.7, 0.9, 1.0])

Trainset = []
for i in range (100):
    data = Data_gen(10)
    Trainset.append(data)
    
    
iteration = 10
RMSE_total = np.zeros(len(lamda))
lamda_all = np.zeros(len(lamda))
for iter in range(len(lamda)):
    print iter
    w = np.zeros(5)
    for i in range (100):
        print i
        Trainset = Data_gen(10)
        w = np.zeros(5)
        [weights,P,RMSE] = learn_oneset(Trainset,0.05,lamda[iter],w)
        RMSE_total[iter] += RMSE
    RMSE_total[iter] /= 100

print RMSE_total    
```
